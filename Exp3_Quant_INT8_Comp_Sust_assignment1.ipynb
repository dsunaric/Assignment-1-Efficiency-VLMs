**## Step 4: Change LLM to Weight-Only 8-bit

```python
%pip install -U bitsandbytes onnx onnxruntime onnxruntime-tools onnxruntime-gpu transformers
```

```python
#%% Install necessary packages (only if not already installed)
# !pip install -q transformers[bitsandbytes] accelerate einops

#%% Imports
import torch
from transformers import LlavaForConditionalGeneration, AutoProcessor

#%% Model setup
model_id = "llava-hf/llava-1.5-7b-hf"

# Load the model with weight-only 8-bit (bitsandbytes)
model = LlavaForConditionalGeneration.from_pretrained(
    model_id,
    device_map="auto",   # automatically place layers on GPU(s)
    torch_dtype=torch.float16,   # <-- weight-only 8-bit
)

# Extract the vision tower
vision_model = model.vision_tower.vision_model.eval()

# Optional: check device
print("Vision tower device:", next(vision_model.parameters()).device)

#%% Processor for images and text
processor = AutoProcessor.from_pretrained(model_id)

#%% Example: forward pass with a dummy image
dummy_input = torch.randn(1, 3, 336, 336, device=next(vision_model.parameters()).device, dtype=torch.float16)

with torch.no_grad():
    embeddings = vision_model(dummy_input)

# bitsandbytes models return BaseModelOutputWithPooling
# embeddings are in embeddings.last_hidden_state or pooled output depending on the model
if hasattr(embeddings, "last_hidden_state"):
    emb_tensor = embeddings.last_hidden_state
else:
    emb_tensor = embeddings[0] if isinstance(embeddings, (tuple, list)) else embeddings

print("✅ Vision tower embeddings shape:", emb_tensor.shape)

#%% Now the model is fully weight-only 8-bit, ready for inference with images
```

```python
import bitsandbytes as bnb
import torch.nn as nn

def quantize_linear8(module):
    for name, child in module.named_children():
        if isinstance(child, nn.Linear):
            # create Linear8bitLt layer
            new_layer = bnb.nn.Linear8bitLt(
                child.in_features,
                child.out_features,
                bias=child.bias is not None
            ).to(child.weight.device, dtype=child.weight.dtype)  # ensure same device + dtype

            # copy weights and bias safely
            with torch.no_grad():  # <-- important
                new_layer.weight.copy_(child.weight)
                if child.bias is not None:
                    new_layer.bias.copy_(child.bias)

            setattr(module, name, new_layer)
        else:
            quantize_linear8(child)
    return module

# Apply INT8 quantization to vision tower
vision_model_int8 = quantize_linear8(model.vision_tower.vision_model)
model.vision_tower.vision_model = vision_model_int8

print("✅ Vision tower quantized to 8-bit with pretrained weights")

```

```python
# Quantize
vision_model_int8 = quantize_linear8(model.vision_tower.vision_model)

# Move to GPU manually
vision_model_int8 = vision_model_int8.to("cuda")

# Replace the original tower
model.vision_tower.vision_model = vision_model_int8

print("✅ Vision tower quantized and moved to:", next(model.vision_tower.vision_model.parameters()).device)
```

```python
import os
import time
import torch
from PIL import Image

# --- Setup ---
image_files = sorted( # sorted so that we avoid randomness
    [f for f in os.listdir(image_directory) if f.endswith('.jpg')]
)
if not image_files:
    raise FileNotFoundError(f"No JPG images found in the directory: {image_directory}")

prompt_text = "Generate a caption for this image"
prompt_template = "USER: <image>\n{prompt_text} ASSISTANT:"

# --- Preprocessing ---
images_data = []
for img_file in image_files[:100]: # edit this number for smaller subset
    image_id = int(os.path.splitext(img_file)[0])
    image_path = os.path.join(image_directory, img_file)
    image = Image.open(image_path).convert("RGB")
    prompt = prompt_template.format(prompt_text=prompt_text)
    inputs = processor(text=prompt, images=image, return_tensors="pt")
    if torch.cuda.is_available():
        inputs = {k: v.to("cuda") for k, v in inputs.items()}
    images_data.append((image_id, inputs))
```

## Step 6: Generating caption with 8-bit model
### Running model with different token settings to compare metrics

```python
import time
from pycocoevalcap.cider.cider import Cider  # or from your local implementation
torch.cuda.empty_cache()
device = next(model.parameters()).device

token_counts = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

results = []

for token_count in token_counts:
    print(f"\n--- Evaluating for max_new_tokens = {token_count} ---")
    generated_captions = {}
    latencies = []

    for image_id, inputs in images_data:
        # Move inputs to correct device
        inputs = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in inputs.items()}

        start_time = time.perf_counter()

        # Generate caption
        generate_ids = model.generate(**inputs, max_new_tokens=token_count)
        generated_text = processor.batch_decode(generate_ids, skip_special_tokens=True)[0]

        end_time = time.perf_counter()
        latency = end_time - start_time
        latencies.append(latency)

        # Extract assistant response
        text = generated_text.split("ASSISTANT:")[-1].strip() if "ASSISTANT:" in generated_text else generated_text.strip()
        generated_captions[image_id] = [text]

    # Metrics
    avg_latency = sum(latencies) / len(latencies)
    throughput = len(latencies) / sum(latencies)

    # Compute CIDEr
    cider_scorer = Cider()
    filtered_gt = {img_id: captions for img_id, captions in ground_truth_captions.items() if img_id in generated_captions}
    cider_score, _ = cider_scorer.compute_score(filtered_gt, generated_captions)

    print(f"Processed {len(latencies)} images | Avg latency: {avg_latency:.3f}s | Throughput: {throughput:.2f} img/s | CIDEr: {cider_score:.3f}")

    results.append({
        "tokens": token_count,
        "avg_latency": avg_latency,
        "throughput": throughput,
        "num_images": len(latencies),
        "cider": cider_score,
    })
```

```python
import json

from datetime import datetime

now = datetime.now()
current_time = now.strftime("%B %d, %H:%M")

# Pretty-print the results array as JSON
print(f"\n=== Final Results from Quant vision tower; {current_time} ===")
print(json.dumps(results, indent=4))
```

```python
# --- After the loop finishes ---
import pandas as pd
import matplotlib.pyplot as plt

# Convert to DataFrame for easier plotting
df = pd.DataFrame(results)

# Create separate plots for each metric
metrics = ["cider", "avg_latency", "throughput"]
titles = {
    "cider": "CIDEr Score vs Max New Tokens",
    "avg_latency": "Average Latency vs Max New Tokens",
    "throughput": "Throughput vs Max New Tokens"
}

for metric in metrics:
    plt.figure(figsize=(6,4))
    plt.plot(df["tokens"], df[metric], marker='o', linestyle='-')
    plt.title(titles[metric])
    plt.xlabel("Max New Tokens")
    plt.ylabel(metric.replace("_", " ").capitalize())
    plt.grid(True)
    plt.show()
```
